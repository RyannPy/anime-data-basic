# Analisis Data Anime

Tujuan : Menganalisis rating anime dari dataset CSV.

## 1.Tampilkan Data

```python
import pandas as pd

df = pd.read_csv("anime.csv")

df
```

## 2. Analisis Data

### 2.1 Filter



#### 2.1.1 Filter Rating
Tujuan : Filter anime berdasarkan rating.

```python
# RATING 0-3
df[df["rating"] < 4]
```

```python
# RATING 4-7
df[(df["rating"] >= 4) & (df["rating"] <= 7)]

```

```python
# RATING 8-10
df[df["rating"] > 7]
```

#### 2.1.2 Filter Tahun Rilis
Tujuan : Filter anime berdasarkan tahun rilis.


```python
# SEBELUM TAHUN 2000
df[df["release"] < 2000]
```

```python
# SETELAH TAHUN 2000
df[df["release"] >= 2000]
```
